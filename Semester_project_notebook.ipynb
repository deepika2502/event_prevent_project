```python
import pandas as pd
import numpy as np
import xml.etree.cElementTree as et
import os

#change path here
path = '/content/drive/MyDrive/Semester Project Deepika, Amir/afp_2021/2021'
#change path here
path_to_write = '/content/drive/MyDrive/Semester Project Deepika, Amir/afp_2021/sentences_dir'

#list directory at level 1 from root
for dir1 in os.listdir(path):
    if not dir1.startswith('.'):

#list directory at level 2 from root
        for dir2 in os.listdir(os.path.join(path,dir1)):
            if not dir2.startswith('.'):

#list directory at level 3 from root
                for dir3 in os.listdir(os.path.join(path,dir1, dir2)):
                    if not dir3.startswith('.'):
#extract sentences
                        tree=et.parse(os.path.join(path,dir1,dir2,dir3))
                        root=tree.getroot()
                        sentences=[]

#iterate through the sentences
                        for sentence in root.iter('p'):
                          sentences.append(sentence.text)

#Convert to dataframe and save to text file
                    Jobs_df = pd.DataFrame(list(zip(sentences)), columns=['sentences'])
                    Jobs_df.to_csv(path_to_write+'/_'+str(dir1)+'_'+str(dir2)+'_'+str(dir3)+'.txt', sep='\t', index=False)
                        
```
